## 3. **플레이 규칙** 
    - 사건 종결 "미해결 엔딩"
        - 시간 제한 내에 맞추어야함
        - 범인 지목을 세 번 틀릴 경우 **패널티** 부여
            - 처음 신뢰도=3
            - 한 번 틀리면: 신뢰도 -1
        - 엔딩 문구: "범인을 밝혀내지 못한 채 사건은 미궁으로 빠졌습니다"
    - 질문 제한
        - 증인에게 질문(5회 제한, hint처럼 사용)
        - 용의자에게 질문(제한 없음)

```python
import json
with open('./../data/case_file1.json', 'r', encoding='utf-8') as f:
    case = json.load(f)
```

```python
# 사건 종결
import timer

def game_over():
    timer = timer()
    total_confidence = 3

    while timer:
        user_input = input('범인 지목: ')
        print('프로파일러가 지목한 범인: ', user_input)
        for truth in case['진실']:
            criminal = truth['진짜 범인']
            print(criminal)

            if criminal == user_input:
                print('[게임 종료] 완벽한 추리입니다. 범인을 밝혀냈습니다.')
                break
            else:
                total_confidence -= 1
                while total_confidence >= 0:
                    if total_confidence == 0:
                        print('[게임 종료] 범인을 밝혀내지 못한 채 사건은 미궁으로 빠졌습니다.')
                        break
                    else:
                        print('[게임 진행] 무고한 사람을 지목하여 신뢰도가 {total_confidence}로 감소합니다. 다시 추리해보세요.')
```
